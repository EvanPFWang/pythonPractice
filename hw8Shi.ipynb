We start with the standard orthogonality relation for the Legendre polynomials on the interval ${[-1, 1]}$:

$$
\int_{-1}^{1} \phi_j(x)\phi_k(x)\,dx =
  \begin{cases}
    0, & j \ne k, \\
    \frac{2}{2j+1}, & j = k.
  \end{cases}
$$

Next, we introduce the linear transformation

$$
t = \frac{1}{2}\Bigl[(b-a)x + (a+b)\Bigr],
$$

which maps ${x \in [-1, 1]}$ to ${t \in [a,b]}$. Its inverse is given by

$$
x = \frac{2t-(a+b)}{b-a},
$$

and the differential transforms as

$$
dt = \frac{b-a}{2}\,dx \quad \text{or} \quad dx = \frac{2}{b-a}\,dt.
$$

Define the new polynomials (with a slight abuse of notation) by

$$
\psi_j(t) = \phi_j\Bigl(\frac{2t-(a+b)}{b-a}\Bigr).
$$

Then the integral over ${[a,b]}$ becomes

$$
\int_a^b \psi_j(t)\psi_k(t)\,dt = \int_{-1}^{1} \phi_j(x)\phi_k(x)\frac{b-a}{2}\,dx.
$$

Using the orthogonality on ${[-1,1]}$, we have

$$
\int_a^b \psi_j(t)\psi_k(t)\,dt = \frac{b-a}{2} \cdot \frac{2}{2j+1}\,\delta_{jk} = \frac{b-a}{2j+1}\,\delta_{jk}.
$$

Thus, we obtain

$$
\int_a^b \phi_j(t)\phi_k(t)\,dt =
  \begin{cases}
    0, & j \ne k, \\
    \frac{b-a}{2j+1}, & j = k.
  \end{cases}
$$

### Analyzing the Recurrence Relation

The Legendre polynomials satisfy the three-term recurrence relation

$$
(j+1)\phi_{j+1}(x) = (2j+1)x\,\phi_j(x) - j\,\phi_{j-1}(x).
$$

When we express this in terms of ${t}$ using

$$
x = \frac{2t-(a+b)}{b-a},
$$

the recurrence becomes

$$
(j+1)\phi_{j+1}\Bigl(\frac{2t-(a+b)}{b-a}\Bigr) = (2j+1)\left(\frac{2t-(a+b)}{b-a}\right)\phi_j\Bigl(\frac{2t-(a+b)}{b-a}\Bigr) - j\,\phi_{j-1}\Bigl(\frac{2t-(a+b)}{b-a}\Bigr).
$$

Defining ${\psi_j(t) = \phi_j\Bigl(\frac{2t-(a+b)}{b-a}\Bigr)}$, the recurrence in the ${t}$-variable is written as

$$
(j+1)\psi_{j+1}(t) = \frac{2j+1}{b-a}\Bigl[2t-(a+b)\Bigr]\psi_j(t) - j\,\psi_{j-1}(t).
$$

Notice that the structure of the recurrence is preserved up to the multiplicative constant ${\frac{1}{b-a}}$ and the shifted argument ${2t-(a+b)}$ in place of ${x}$.

### Conclusion

The change of variable not only transforms the integration limits and the measure (introducing the factor ${\frac{b-a}{2}}$) but also leads to a modified recurrence relation in the ${t}$-variable. Together, these ensure that the orthogonality relation becomes

$$
\int_a^b \phi_j(t)\phi_k(t)\,dt =
  \begin{cases}
    0, & j \ne k, \\
    \frac{b-a}{2j+1}, & j = k,
  \end{cases}
$$

which is the desired result.
