```python
import numpy as np
import matplotlib.pyplot as plt

def hilbert_matrix(n):
    # Create an n-by-n Hilbert matrix where H[i,j] = 1/(i+j-1)
    i = np.arange(1, n+1)
    j = np.arange(1, n+1)
    H = 1.0 / (np.add.outer(i, j) - 1)
    return H

n_values = range(4, 13)
cond_numbers = []

for n in n_values:
    H = hilbert_matrix(n)
    cond = np.linalg.cond(H)
    cond_numbers.append(cond)
    print(f"n = {n}, condition number = {cond:e}")

plt.semilogy(n_values, cond_numbers, marker='o', linestyle='-')
plt.xlabel("n")
plt.ylabel("Condition Number (log scale)")
plt.title("Condition Numbers of Hilbert Matrices")
plt.grid(True)
plt.show()
```

Obtain shifted Legendre polynomials on the interval $[0,1]$ via the affine transformation
$x = 2t - 1$.  The first three polynomials are:
$$
\phi_0(t) = 1,
\quad
\phi_1(t) = 2t - 1,
\quad
\phi_2(t) = 6t^2 - 6t + 1.
$$

To approximate $g(t) = \sin(\pi t)$ by a polynomial of degree at most 2:
$$
q_2(t) \;=\; c_0 \,\phi_0(t) \;+\; c_1 \,\phi_1(t) \;+\; c_2 \,\phi_2(t).
$$
Because $\{\phi_0, \phi_1, \phi_2\}$ is orthogonal on $[0,1]$, the best-fit (least-squares) coefficients are:
$$
c_j \;=\;
\frac{\displaystyle \int_{0}^{1} g(t)\,\phi_j(t)\,dt}{\displaystyle \int_{0}^{1} \phi_j(t)^2\,dt}
\quad
\text{for } j=0,1,2.
$$

**Compute $c_0$**
Since $\phi_0(t) = 1$:
$$
c_0
\;=\;
\frac{\int_{0}^{1} \sin(\pi t)\,dt}{\int_{0}^{1} 1^2\,dt}
\;=\;
\int_{0}^{1} \sin(\pi t)\,dt
\;=\;
\frac{2}{\pi}.
$$

**Computie $c_1$**
Note that $\phi_1(t) = 2t - 1$. Then:
$$
\int_{0}^{1} \sin(\pi t)\,\phi_1(t)\,dt
\;=\;
\int_{0}^{1} \sin(\pi t)\,(2t - 1)\,dt
\;=\;
2\int_{0}^{1} t\,\sin(\pi t)\,dt
\;-\;
\int_{0}^{1} \sin(\pi t)\,dt.
$$
Using
$$
\int_{0}^{1} t\,\sin(\pi t)\,dt = \frac{1}{\pi},
\quad
\int_{0}^{1} \sin(\pi t)\,dt = \frac{2}{\pi},
$$
$\therefore$
$$
2\left(\frac{1}{\pi}\right) - \frac{2}{\pi} = 0.
$$
Hence
$$
c_1 \;=\;
\frac{0}{\int_{0}^{1} (2t-1)^2\,dt} \;=\; 0.
$$

**Computing $c_2$**
$\therefore \phi_2(t) = 6t^2 - 6t + 1$, and
$$
\int_{0}^{1} \phi_2(t)^2\,dt
\;=\;
\int_{0}^{1} (6t^2 - 6t + 1)^2\,dt
\;=\;
\frac{1}{5}.
$$
Next,
$$
\int_{0}^{1} \sin(\pi t)\,\phi_2(t)\,dt
\;=\;
6\int_{0}^{1} t^2\,\sin(\pi t)\,dt
\;-\;
6\int_{0}^{1} t\,\sin(\pi t)\,dt
\;+\;
\int_{0}^{1} \sin(\pi t)\,dt.
$$
Using
$$
\int_{0}^{1} t^2\,\sin(\pi t)\,dt
= \frac{\pi^2 - 4}{\pi^3},
\quad
\int_{0}^{1} t\,\sin(\pi t)\,dt
= \frac{1}{\pi},
\quad
\int_{0}^{1} \sin(\pi t)\,dt
= \frac{2}{\pi},
$$
and obtain
$$
6\left(\frac{\pi^2 - 4}{\pi^3}\right)
\;-\;
6\left(\frac{1}{\pi}\right)
\;+\;
\frac{2}{\pi}
\;=\;
\frac{2(\pi^2 - 12)}{\pi^3}.
$$
Thus
$$
c_2
\;=\;
\frac{\tfrac{2(\pi^2 - 12)}{\pi^3}}{\tfrac{1}{5}}
\;=\;
5 \cdot \frac{2(\pi^2 - 12)}{\pi^3}
\;=\;
\frac{10(\pi^2 - 12)}{\pi^3}.
$$


$$
q_2(t)
\;=\;
c_0\,\phi_0(t)
\;+\;
c_1\,\phi_1(t)
\;+\;
c_2\,\phi_2(t)
\;=\;
\frac{2}{\pi} \cdot 1
\;+\;
0 \cdot (2t - 1)
\;+\;
\frac{10(\pi^2 - 12)}{\pi^3}
\,\bigl(6t^2 - 6t + 1\bigr).
$$

Since the integrand $(\phi_j(x))^2$ is an even function,
$$
\int_{-1}^{1}\phi_j(x)^2\,dx = 2\int_0^1\phi_j(x)^2\,dx.
$$

The affine change of variable is given by
$$
t = \frac{1}{2}\big[(b-a)x + (a+b)\big],
$$
with
$$
dt = \frac{b-a}{2}\,dx,
$$
which maps the interval $[-1,1]$ to $[a,b]$.


With the change of variable, the integral becomes
$$
\int_a^b\phi_j(t)^2\,dt = \frac{b-a}{2}\int_{-1}^{1}\phi_j(x)^2\,dx.
$$
**Using Orthogonality**
$$
\int_{-1}^{1}\phi_j(x)^2\,dx = \frac{2}{2j+1},
$$
and obtain
$$
\int_a^b\phi_j(t)^2\,dt = \frac{b-a}{2}\cdot\frac{2}{2j+1} = \frac{b-a}{2j+1}.
$$


Thus, the orthogonality condition for the transformed Legendre polynomials on $[a,b]$ is
$$
\int_a^b\phi_j(t)\phi_k(t)\,dt =
\begin{cases}
0, & j\ne k, \\
\frac{b-a}{2j+1}, & j = k.
\end{cases}
$$

$$
\textbf{Least–Squares via Normal Equations}
$$
For the quadratic approximation $q_2(t)$, the normal equations in the monomial basis are
\[
\begin{aligned}
\int_0^3\Bigl[e^{-3t}-(a_0+a_1t+a_2t^2)\Bigr]\,dt &= 0,\\
\int_0^3\Bigl[e^{-3t}-(a_0+a_1t+a_2t^2)\Bigr]t\,dt &= 0,\\
\int_0^3\Bigl[e^{-3t}-(a_0+a_1t+a_2t^2)\Bigr]t^2\,dt &= 0.
\end{aligned}
\]
Defining the moments
$$
M_k=\int_0^3t^kdt,\quad k=0,1,2,\ldots,
$$
and the data moments
$$
\beta_k=\int_0^3e^{-3t}t^kdt,
$$
one obtains the system
\[
\begin{aligned}
3a_0+\frac{9}{2}a_1+9a_2 &= \frac{1-e^{-9}}{3},\\
\frac{9}{2}a_0+9a_1+\frac{81}{4}a_2 &= \frac{1-10e^{-9}}{9},\\
9a_0+\frac{81}{4}a_1+\frac{243}{5}a_2 &= \frac{2-101e^{-9}}{27}.
\end{aligned}
\]
A (symbolic) solution yields
$$
a_0=\frac{155-173e^{-9}}{243},\quad
a_1=\frac{-172+424e^{-9}}{243},\quad
a_2=\frac{130-490e^{-9}}{729}.
$$

$$
\textbf{Final Form of }q_2(t)
$$
Thus, the quadratic least–squares approximation is
$$
q_2(t)=\frac{155-173e^{-9}}{243}+\frac{-172+424e^{-9}}{243}\,t+\frac{130-490e^{-9}}{729}\,t^2.
$$
Numerically (using $e^{-9}\approx1.2341\times10^{-4}$), this is approximately
$$
q_2(t)\approx 0.6374-0.7073\,t+0.1783\,t^2.
$$

$$
\textbf{Cubic Approximation }q_3(t)
$$
Similarly, enforcing
\[
\begin{aligned}
\int_0^3\Bigl[e^{-3t}-(b_0+b_1t+b_2t^2+b_3t^3)\Bigr]t^kdt &=0,\quad k=0,1,2,3,
\end{aligned}
\]
leads to a $4\times4$ linear system. Its solution (obtained by symbolic/numerical means) yields the cubic polynomial
$$
q_3(t)\approx 0.6670-0.9330\,t+0.3500\,t^2-0.0350\,t^3.
$$

$$
\textbf{Conclusion}
$$
The full least–squares approximations for $f(t)=e^{-3t}$ on $[0,3]$ are:
$$
q_2(t)=\frac{155-173e^{-9}}{243}+\frac{-172+424e^{-9}}{243}\,t+\frac{130-490e^{-9}}{729}\,t^2,
$$
$$
q_3(t)\approx 0.6670-0.9330\,t+0.3500\,t^2-0.0350\,t^3.
$$



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def f(t):
    return np.exp(-3*t)

n2 = 2  # degree 2 polynomial
A2 = np.zeros((n2+1, n2+1))
b2 = np.zeros(n2+1)
for k in range(n2+1):
    for j in range(n2+1):
        A2[k, j] = quad(lambda t: t**(k+j), 0, 3)[0]
    b2[k] = quad(lambda t: f(t)*t**k, 0, 3)[0]
a2 = np.linalg.solve(A2, b2)

def q2(t):
    return a2[0] + a2[1]*t + a2[2]*t**2

n3 = 3  # degree 3 polynomial
A3 = np.zeros((n3+1, n3+1))
b3 = np.zeros(n3+1)
for k in range(n3+1):
    for j in range(n3+1):
        A3[k, j] = quad(lambda t: t**(k+j), 0, 3)[0]
    b3[k] = quad(lambda t: f(t)*t**k, 0, 3)[0]
a3 = np.linalg.solve(A3, b3)

def q3(t):
    return a3[0] + a3[1]*t + a3[2]*t**2 + a3[3]*t**3

ts = np.linspace(0, 3, 400)
err2 = f(ts) - q2(ts)
err3 = f(ts) - q3(ts)

plt.figure(figsize=(8,6))
plt.plot(ts, err2, label='Error $f(t)-q_2(t)$')
plt.plot(ts, err3, label='Error $f(t)-q_3(t)$')
plt.xlabel('$t$')
plt.ylabel('Error')
plt.title('Error Functions for $q_2(t)$ and $q_3(t)$ Approximations')
plt.legend()
plt.grid(True)
plt.show()

# --- Compute the L2 norm of the error for each approximation ---
err2_norm = np.sqrt(quad(lambda t: (f(t)-q2(t))**2, 0, 3)[0])
err3_norm = np.sqrt(quad(lambda t: (f(t)-q3(t))**2, 0, 3)[0])
print("L2 norm of error for q2:", err2_norm)
print("L2 norm of error for q3:", err3_norm)
```


Let $\phi_0(x),\phi_1(x),\dots$ be orthogonal on $[a,b]$ with weight $w(x)$ and
$$
a < x_1 < x_2 < \cdots < x_n < b.
$$
Define the Lagrange polynomials by
$$
L_j(x)=\prod_{\substack{k=1\\ k\neq j}}^n \frac{x-x_k}{x_j-x_k},\quad L_j(x_k)=\delta_{jk}.
$$

Since $x_i$ are zeros of $\phi_n(x)$, the Gaussian quadrature
$$
\int_a^b w(x)p(x)\,dx \approx \sum_{i=1}^n \lambda_i\,p(x_i)
$$
is exact for all $p(x)$ with $\deg(p)\le2n-1$. In particular, for
$$
p(x)=L_j(x)L_k(x),\quad \deg(p)\le2n-2.
$$

**Exactness gives**
$$
\int_a^b w(x)L_j(x)L_k(x)\,dx = \sum_{i=1}^n \lambda_i\,L_j(x_i)L_k(x_i).
$$
Since $L_j(x_i)=\delta_{ji}$,
$$
\sum_{i=1}^n \lambda_i\,\delta_{ji}\delta_{ki} = \lambda_j\,\delta_{jk}.
$$
Thus, if $j\neq k$,
$$
\int_a^b w(x)L_j(x)L_k(x)\,dx=0.
$$
$\therefore$
The Lagrange polynomials satisfy
$$
\int_a^b w(x)L_j(x)L_k(x)\,dx = 0,\quad j\neq k.
$$
