```python
class UnderflowException(Exception):
    pass

class binHeapKeyed:
    def __init__(self,capacity, key =   None):
        self.size = 0
        arr =   []*(capacity+1)
        self.key    =   key #if bin_heap holds iter

    def insert(self, x):
        if self.key is  None:
            if  hasattr(x,"__getitem__"):
                raise   TypeError("Iterable elem, expected non-iter from key = None")
            comp_val  =   x
        else:
            if  not  hasattr(x,"__getitem__"):
                raise TypeError("non-Iterable elem, expected iter from key  !=  None")
            comp_val    =   x[self.key]
        
        if self.size    ==  len(self.arr)-1:
            self.enlarge(len(self.arr)*2+1)
        hole_key    =   self.size    +   1
        while   hole_key>1  and comp_val < self.get_compare_value(self.arr[hole_key//2]):
            self.arr[hole_key]  =   self.arr[hole_key//2]
            hole_key    //= 2
        self.arr[hole_key]  =   comp_val
        self.size +=1
        
    
    def get_compare_value (self,x):
        return  x[self.key] if  self.key is not None else   x

    def enlarge(self, newSize):
        old = self.arr
        self.arr = [None] * newSize
        for i in range(len(old)):
            self.arr[i] = old[i]

    def findMin(self):
        if self.is_empt():
            raise UnderflowException()
        return self.arr[1]

    def deleteMin(self):
        if self.is_empt():
            raise   UnderflowException()
        min_item    =   self.findMin()
        self.arr[1] =   self.arr[self.size]
        self.size -= 1
        self.percolateDown(1)
        return min_item
    

    def percolateDown(self, ind):
        temp = self.arr[ind]
        while   ind   *   2 <   self.size:
            child_ind = ind *   2
            if child_ind    !=  self.size   and self.get_compare_value(self.arr[child_ind+1]) < self.get_compare_value(self.arr[child_ind]):
                child_ind = child_ind + 1
            if  self.get_compare_value(self.arr[child_ind]) <   self.get_compare_value(temp):
                self.arr[ind]   =   self.arr[child_ind]
            else:
                break
        self.arr[ind] = temp

    def is_empt(self):
        return self.size == 0

    def makEmpty(self):
        self.size = 0
        self.arr = [None] * len(self.arr)
```
