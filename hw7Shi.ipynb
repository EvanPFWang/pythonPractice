For $n+1$ equally spaced points $x_0, x_1, \dots, x_n$, 
the classical Newton interpolation polynomial can be written in terms of 
\emph{divided differences} as
$$
\
p_n(x) 
\;=\; 
\sum_{j=0}^n 
f[x_0, x_1, \dots, x_j]\;\bigl(x - x_0\bigr)\bigl(x - x_1\bigr)\cdots\bigl(x - x_{j-1}\bigr).
\
$$
Here $f[x_0,\dots,x_j]$ denotes the $j$th divided difference of $f$.



From the proof by induction on the order of forward differences, one establishes
$$
\
f[x_0, x_1, \dots, x_j] 
\;=\;
\frac{\Delta^j f(x_0)}{j!\,h^j},
\
$$
where $\Delta^j f(x_0)$ is the $j$th forward difference of $f$ evaluated at $x_0$.


Because $x_i = x_0 + i\,h$, each factor is
$$
\
x - x_k \;=\; \bigl(x_0 + s\,h\bigr) - \bigl(x_0 + k\,h\bigr) 
\;=\; h\,\bigl(s - k\bigr),
\
$$
so
$$
\
(x - x_0)(x - x_1)\dots(x - x_{j-1}) 
\;=\; 
h^j \,\bigl(s - 0\bigr)\bigl(s - 1\bigr)\dots\bigl(s - (j-1)\bigr).
\
$$
Factor out $j!$ in the denominator to write
$$
\
\bigl(s - 0\bigr)\bigl(s - 1\bigr)\dots\bigl(s - (j-1)\bigr)
\;=\;
j! \,\binom{s}{j},
\
$$
where 
$$
\
\binom{s}{j} \;=\; \frac{s(s-1)\cdots(s-j+1)}{j!}.
\
$$

Substituting $f[x_0,\dots,x_j] = \tfrac{\Delta^j f(x_0)}{j!\,h^j}$ and the factorization of $(x - x_0)\dots(x - x_{j-1})$, we get
$$
\
p_n(x)
\;=\;
\sum_{j=0}^n 
\frac{\Delta^j f(x_0)}{j!\,h^j}
\;\Bigl[h^j \,j!\,\binom{s}{j}\Bigr]
\;=\;
\sum_{j=0}^n 
\binom{s}{j}\,\Delta^j f(x_0).
\
$$
Since $s = \tfrac{x - x_0}{h}$, we have precisely
$$
\
p_n(x) 
\;=\; 
\sum_{j=0}^n \binom{s}{j}\,\Delta^j f(x_0),
\
$$

## Part (a): Cubic Polynomial with Nodes at $x_0 = 1, x_1 = 3$


The interpolant exactly matches $f(x)$ and $f'(x)$ at $x_0 = 1$ and $x_1 = 3$.
However, $\cosh(x)$ grows rapidly on $[1,3]$, so the error between the endpoints increases, typically reaching about $10^{-3}$ in the middle of the interval.
Without additional nodes or higher-order derivative constraints, the cubic struggles to approximate $f(x)$ in the region away from $x_0$ and $x_1$.



## Part (b): Quintic Polynomial with Nodes at $x_0 = 1$, $x_1 = 2$, $x_2 = 3$

A quintic polynomial (degree $5$) is constructed to satisfy
$$
p(x_i) = f(x_i)
\quad\text{and}\quad
p'(x_i) = f'(x_i)
\quad
\text{for } i = 0,1,2.
$$
This gives $6$ total conditions (since $3$ points $\times$ $2$ conditions per point $= 6$), exactly enough to fix a degree-$5$ polynomial.

 - The extra interpolation node at $x_1 = 2$ and the higher degree give more flexibility, so the polynomial generally fits $\cosh(x)$ better on $[1,3]$.
 - The maximum error typically drops below that of the cubic, often by one or more orders of magnitude.
 - Plots of the error (using a semilogy scale) confirm a smaller peak error compared to part (a).

## Part (c): Higher-Order Polynomial with Second-Derivative Matching

Here, the interpolant is constructed to match
$$
p(x_i) = f(x_i),
\quad
p'(x_i) = f'(x_i),
\quad
p''(x_i) = f''(x_i)
$$
for selected nodes (e.g.\ $x_0 = 1$ and $x_1 = 3$). Each second-derivative match adds another condition, increasing the polynomial's degree or reducing the number of nodes needed.

 - Enforcing second-derivative conditions can significantly improve the local accuracy near the chosen points.
 - The overall error across $[1,3]$ is often smaller than in parts (a) and (b), though $\cosh(x)$ still grows quickly.
 - In practice, combining additional nodes with higher-order derivative constraints yields the best balance of local and global accuracy.

## Overall Comparison

- **Cubic (Part a):** Matches $f$ and $f'$ at two nodes; simplest approach but higher error in the interior.
- **Quintic (Part b):** Matches $f$ and $f'$ at three nodes; lower overall error due to the extra node and higher degree.
- **Second-Derivative (Part c):** Improves local accuracy further by matching $f''$; can reduce the error near boundary nodes or across the interval, depending on the polynomial degree.

\
\begin{array}{rcl}
\textbf{(a)}\quad &:& \text{Let } P(x)=a+bx+cx^2 \text{ be the quadratic interpolant satisfying:}\$$
P(0)=a &=& f(0)=1,\$$
P'(0)=b &=& f'(0)=-1,\$$
P(1)=a+b+c &=& f(1)=2.
\end{array}
\
$$

Since \(a=1\) and \(b=-1\), the third condition becomes
$$
\
1 - 1 + c = 2 \quad \Longrightarrow \quad c=2.
\
$$
Thus, the interpolant is
$$
\
P(x)=1-x+2x^2.
\
$$
Its derivative is
$$
\
P'(x)=-1+4x.
\
$$
Setting \(P'(x)=0\) yields
$$
\
-1+4x=0 \quad \Longrightarrow \quad x=\frac{1}{4}.
\
$$
Therefore, the approximate minimizer is
$$
\
x^*=\frac{1}{4}.
\
$$

$$
\
\textbf{(b)}\quad \text{Since } P''(x)=4>0 \text{ for all } x, \text{ the quadratic } P(x) \text{ is strictly convex. Hence, it has a unique minimum at } x^*=\frac{1}{4}.
\
$$
Moreover, clearly \(0<\frac{1}{4}<1\).

Now, consider the function $\(f(x)\)$ itself. The conditions
$$
\
f(0)<f(1) \quad \text{and} \quad f'(0)<0
\
$$
imply that $(f)$ is initially decreasing at $(x=0)$ and, since $(f(0)<f(1))$, it must eventually increase on $([0,1])$. Under the common assumption (in weak line search methods) that $(f)$ is unimodal on $([0,1])$, there exists a unique point $(x^*$in$(0,1))$ where $(f'(x^*)=0)$; this point is the unique minimizer of (f) on ([0,1]).

Thus, both the quadratic interpolant and, under the unimodality assumption, the function \(f\) itself have a unique minimum satisfying
$$
\
0<x^*<1.
\
$$
