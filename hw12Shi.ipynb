**3**

Since backward Euler approximates the derivative at \(t_{i+1}\) by
$$
y'(t_{i+1}) \approx \frac{y(t_{i+1}) - y(t_i)}{h},
$$
the local truncation error \(d_i\) is defined by
$$
d_i
= \frac{y(t_{i+1}) - y(t_i)}{h} \;-\; f\bigl(t_{i+1},\,y(t_{i+1})\bigr).
$$

$$
0 \;=\; \frac{y_{i+1} - y_i}{h} \;-\; f\bigl(t_{i+1},\,y_{i+1}\bigr).
$$


$$
\therefore d_i - 0
= \frac{y(t_{i+1}) - y(t_i)}{h}
  \;-\;\frac{y_{i+1} - y_i}{h}
  \;-\;\bigl[f\bigl(t_{i+1},y(t_{i+1})\bigr) - f\bigl(t_{i+1},y_{i+1}\bigr)\bigr],
$$
$$
(e_i = y(t_i) - y_i) \implies d_i
= \frac{e_{i+1} - e_i}{h}
  - \bigl[f(t_{i+1},y(t_{i+1})) - f(t_{i+1},y_{i+1})\bigr].
$$

Assume f is Lipschitz in y with constant L
$$
\therefore \bigl|f(t,y) - f(t,y_*)\bigr| \le L\,|y - y_*|.
$$

$$
\bigl|e_{i+1}\bigr|
= \Bigl|\,e_i + h\bigl[f(t_{i+1},y(t_{i+1})) - f(t_{i+1},y_{i+1})\bigr] + h\,d_i\Bigr|
\le |e_i| + h\,L\,|e_{i+1}| + h\,|d_i|.
$$
Rearranging (for \(hL<1\)) gives
$$
(1 - hL)\,|e_{i+1}|
\le |e_i| + h\,|d_i|
\quad\Longrightarrow \quad
|e_{i+1}| \le \frac{|e_i| + h\,|d_i|}{1 - hL}.
$$



For global err bound $\implies$ Ass. bounded true soln. second derivative,
$$
\bigl|y''(t)\bigr| \le M,
$$
then local trunc error satisfies \(\,|d_i|\le \tfrac{Mh}{2}\,\).
Iterating the inequality yields the global error estimate
$$
|e_i| \le \frac{Mh}{2L}\,\bigl(e^{L(t_i - a)} - 1\bigr),
\quad i = 0,1,\dots,N.
$$


**6**
(a) Explicit trapezoid method for $y' = f(y)$.
$$
  y(t_{i+1}) = y_i + h\,f(t_i,y_i) + \frac{h^2}{2}\,f_y(t_i,y_i)\,f(t_i,y_i) + \mathcal{O}(h^3).
$$
    If
$$
  f\bigl(y_i + h\,f(t_i,y_i)\bigr)
  = f(t_i,y_i) + h\,f_y(t_i,y_i)\,f(t_i,y_i) + \mathcal{O}(h^2).
$$

$$
\therefore
$$
The explicit trap step is
$$
  y_{i+1} = y_i + \frac{h}{2}\Bigl(f(t_i,y_i) + f\bigl(y_i + h\,f(t_i,y_i)\bigr)\Bigr)
  = y_i + \frac{h}{2}\Bigl(2\,f(t_i,y_i) + h\,f_y(t_i,y_i)\,f(t_i,y_i) + \mathcal{O}(h^2)\Bigr)
  = y_i + h\,f(t_i,y_i) + \frac{h^2}{2}\,f_y(t_i,y_i)\,f(t_i,y_i) + \mathcal{O}(h^3).
$$
Matching Taylor expansion up to $h^2$ shows local trunc error of $\mathcal{O}(h^3)$  also second order accurate.

(b) Explicit midpoint method for $y' = f(y)$
$$
  y(t_i + h) = y_i + h\,f(t_i,y_i) + \frac{h^2}{2}\,f_y(t_i,y_i)\,f(t_i,y_i) + \mathcal{O}(h^3).
$$

If
$$
  f\!\Bigl(y_i + \tfrac{h}{2}\,f(t_i,y_i)\Bigr)
  = f(t_i,y_i) + \frac{h}{2}\,f_y(t_i,y_i)\,f(t_i,y_i) + \mathcal{O}(h^2).
$$
$$
\therefore
$$
The explicit midpoint step is
$$
  y_{i+1} = y_i + h\,f\!\Bigl(y_i + \tfrac{h}{2}\,f(t_i,y_i)\Bigr)=
  y_{i+1}
  = y_i + h\Bigl(f(t_i,y_i) + \tfrac{h}{2}\,f_y(t_i,y_i)\,f(t_i,y_i) + \mathcal{O}(h^2)\Bigr)
  = y_i + h\,f(t_i,y_i) + \frac{h^2}{2}\,f_y(t_i,y_i)\,f(t_i,y_i) + \mathcal{O}(h^3).
$$
Matching Taylor expansion up to $h^2$ shows local trunc error of $\mathcal{O}(h^3)$  also second order accurate.


**12**
Interpolate through $(t_{i-1},\,f_{i-1}),\quad (t_i,\,f_i),\quad (t_{i+1},\,f_{i+1}).$


The $s$-step is obtained by interpolating $f$ through previous $s$ points plus the next one,
$$
t_{i+1},\;t_i,\;\dots,\;t_{i+1-s},
$$
and discretely integrating from $t_i$ to $t_{i+1}$.  With $(s=2)+1$ nodes spaced by $h$, the order of accuracy $q = 2 + 1 = 3$

Fit a quadratic through the points
$$(t_{i-1},f_{i-1}),\;(t_i,f_i),\;(t_{i+1},f_{i+1}),$$
then integrate from $t_i$ to $t_{i+1}$.

Let
$$
t_{i-1} = t_0 + (i-1)h,\quad
t_i     = t_0 + i\,h,\quad
t_{i+1} = t_0 + (i+1)h.
$$
Normalize t to
$$
x = \frac{t - t_i}{h},
$$
so $x=-1,0,1$ from $t_{i-1},t_i,t_{i+1}$ respectively


Define the Lagrange basis polyn in $x$:
$$
L_{i+1}(x) = \frac{x(x+1)}{2},\quad
L_i(x)     = 1 - x^2,\quad
L_{i-1}(x) = \frac{x(x-1)}{2}.
$$
$L_j(x_k)=\delta_{jk}$ for $x_k\in\{-1,0,1\}$.

$$
P(t) \;=\; f_{i+1}\,L_{i+1}(x)\;+\;f_i\,L_i(x)\;+\;f_{i-1}\,L_{i-1}(x).
$$

Approx
$$
\int_{t_i}^{t_{i+1}} f(t)\,dt
\;\approx\;
\int_{t_i}^{t_{i+1}} P(t)\,dt
\;=\;
h \int_{0}^{1} P(x)\,dx.
$$
Compute moments:
$$
\int_0^1 L_{i+1}(x)\,dx = \frac{5}{12},\quad
\int_0^1 L_i(x)\,dx     = \frac{8}{12},\quad
\int_0^1 L_{i-1}(x)\,dx = -\frac{1}{12}.
$$

$$
\therefore
\int_{t_i}^{t_{i+1}} P(t)\,dt
= \frac{h}{12}\bigl(5\,f_{i+1} + 8\,f_i - f_{i-1}\bigr).
$$

yields:
$$
y_{i+1}
= y_i
+ \frac{h}{12}\Bigl(
      5\,f\bigl(t_{i+1},y_{i+1}\bigr)
    + 8\,f\bigl(t_i,y_i\bigr)
    -  f\bigl(t_{i-1},y_{i-1}\bigr)
\Bigr).=\; y_i \;+\;\frac{h}{12}\,\bigl(5\,f_{i+1} + 8\,f_i - f_{i-1}\bigr).
$$
